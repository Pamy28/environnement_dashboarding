Code pays suite à traduction

```python
#FR
import pycountry
def get_country_code_fr(country):
    try:
        return pycountry.countries.get(name=country).alpha_2
    except AttributeError:
        return float('NaN')
```

```python
dfrich["Country_Code"] = dfrich["Country"].apply(get_country_code_fr)
```
