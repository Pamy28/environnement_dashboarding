```python
import pandas as pd
dfCO2 = pd.read_excel(repertoire + '/CO2_habitant.xlsx')
dfEnergie = pd.read_excel(repertoire + '/EnergieConsumption.xlsx')
dfrich = pd.read_excel(repertoire + '/niveau_richesse.xlsx')
dfpop = pd.read_excel(repertoire + '/population.xlsx')
dfevol = pd.read_excel(repertoire + '/temp_evol.xlsx')
dfevolmonth = pd.read_excel(repertoire + '/temp_ev_month.xlsx')
```

*Uniformisation* des fichiers (colonnes Year en lignes)

```python
dfCO2_melted = dfCO2.melt(id_vars=['Country Name', 'Indicator Name'], var_name='Year', value_name='CO2_level')
dfCO2_melted[dfCO2_melted['Country Name']=='France']
```

```python
dfevolmonth_melted = dfevolmonth.melt(id_vars=['Unnamed: 0','Country','Months','Element'], var_name='Year', value_name='Temp_values')
dfevolmonth_melted[dfevolmonth_melted['Country']=='France']
```

Mettre les années sans "Y"

```python
dfevolmonth_melted["Year"] = dfevolmonth_melted["Year"].str.replace('Y', '')
```
